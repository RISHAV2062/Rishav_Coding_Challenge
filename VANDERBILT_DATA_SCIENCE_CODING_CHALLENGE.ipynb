```python
#Name: RISHAV CHANDRA ACHARYA
#Email: rishav.c.acharya@vanderbilt.edu
#VUNetID: acharyrc

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/california_housing_test.csv')
#The dataset includes information about California Housing across various indicators
df


```

Cleaning the Data


```python
#Removing duplicated rows

df_cleaned = df.drop_duplicates()

#Handling missing data: filling missing values with the mean of the column

df_cleaned = df.fillna(df.mean())


# Renaming Coulums for better readability
df_cleaned = df_cleaned.rename(columns={
    'housing_median_age': 'Housing Median Age',
    'total_rooms': 'Total Rooms',
    'total_bedrooms': 'Total Bedrooms',
    'median_income': 'Median Income',
    'median_house_value': 'Median House Value'

})
#Normalizing 'Median Income'
#Using Z-score normalization
df_cleaned['Median Income'] = (df_cleaned['Median Income']-df_cleaned['Median Income'].mean())/ df_cleaned['Median Income'].std()

#Removing Outliers
#Removing rows where Median Income is more than 3 standard deviations from the mean
df_cleaned = df_cleaned[(df_cleaned['Median Income']<3) & df_cleaned['Median Income']>-3]
df_cleaned.head()


```

Separating features (X) and target Variable (Y)

```python
#Features: All target variables except 'Median House Value'
X = df_cleaned.drop(columns=['Median House Value'])
X

```

```python
# Target variable: Median House Value

y = df_cleaned['Median House Value']
y
```

```python
#Importing libraries
from sklearn.model_selection import train_test_split

#Splitting the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)

X_train # Training set contains 80 % of the data
```

```python
X_test # Testing set contains 20% of the data
```

```python
# Model Building : Linear Regression

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

LINEAR REGRESSION MODEL


```python
#APPLYING THE MODEL TO MAKE A PREDICTION

y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
print(y_lr_train_pred, y_lr_test_pred) # Predicted Values
```

```python
#Evaluating Model Performance

from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred) # For the training set
lr_train_r2 = r2_score(y_train, y_lr_train_pred) # For the training set

#For the testing set
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

#r2 is the coefficient of determination; regression score function.

#Printing evaluation metrics
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)

#Printing the results in tabular form
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

```python
from sklearn.ensemble import RandomForestRegressor

#Training the model using the RandomForestRegresssor ALgorithm

rf = RandomForestRegressor(max_depth = 2, random_state = 32)
rf.fit(X_train, y_train)
RandomForestRegressor(max_depth=2, random_state = 32)

#Applying the model to Make Predictions for the testing and training sets
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)

#Evaluating Model Performance
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_lr_train_pred)
rf_train_r2 = r2_score(y_train, y_lr_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)

#Printing the results in tabular form
rf_results = pd.DataFrame(['RandomForest Regressor', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2' , 'Test MSE', 'Test R2']
rf_results


```
